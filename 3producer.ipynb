```python
import pymongo
import pandas as pd
from confluent_kafka import Producer
import json
```

```python
Case='<Case File Path>Region='<Region File Path>TimeProvince='<Time Province File Path>'
```

```python
case=pd.read_csv(Case)
region=pd.read_csv(Region)
time_province=pd.read_csv(TimeProvince)
```

```python
case_1=case.to_json(orient='records')
region_1=region.to_json(orient='records')
time_province_1=time_province.to_json(orient='records')
```

```python
case_3=json.loads(case_1)
region_3=json.loads(region_1)
time_province_3=json.loads(time_province_1)
```

```python
def login_info():
    info={
        'bootstrap.servers':'KAFKA SERVER ADDRESS',
        'security.protocol':'SASL_SSL',
        'sasl.mechanism':'PLAIN',
        'sasl.username':'SERVER USERNAME',
        'sasl.password':'SERVER PASSWORD',
        'partitioner':'consistent_random'

    }
    return info
```

```python
def delivery_status(err,msg):
    if err:
        print(f'Message failed to deliver due to the following error {str(err)}')
    else:
        print(f'The Message has been delivered successfully {msg.value()}. Partition-->{msg.partition()} at the Offset--->{msg.offset()}')
        print(f'The following messages are pushed to Topic--->{msg.topic()}')
```

```python
p=Producer(login_info())
```

```python
print(p.list_topics().topics)
```

```python
try:
    for j in range(0,len(region_3)):
        msg=region_3[j]
        p.poll(1.0)
        p.produce(topic='Region',value=str(msg).encode('utf-8'),on_delivery=delivery_status)
except er as Exception:
    print('The following error was failed due to the error-->',er)
print('Flushing records')

p.flush()
```

```python
try:
    for j in range(0,len(case_3)):
        msg=case_3[j]
        p.poll(1.0)
        p.produce(topic='Case',value=str(msg).encode('utf-8'),on_delivery=delivery_status)
except er as Exception:
    print('The following error was failed due to the error-->',er)
print('Flushing records')

p.flush()
```

```python
try:
    for j in range(0,len(time_province_3)):
        msg=time_province_3[j]
        p.poll(1.0)
        p.produce(topic='TimeProvince',value=str(msg).encode('utf-8'),on_delivery=delivery_status)
except er as Exception:
    print('The following error was failed due to the error-->',er)
print('Flushing records')

p.flush()
```
